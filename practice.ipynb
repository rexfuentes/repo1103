#### while loop

### "repeating action until condition is met"

```python
error = 90.0

while error > 1:
    error = error/4
    print(error)
```

```python
offset = 8
while offset != 0:
    print('correcting...')
    offset = offset - 1
    print(offset)
```

```python
offset = -6
while offset != 0:
    print("correcting...")
    if offset > 0:
        offset = offset - 1
    else:
        offset = offset + 1
    print(offset)
```

### for loop

#### for var in seq:
#### expression
### "for each var in seq, execute the expression"


```python
fam = [1.73, 1.68, 1.71, 1.89]

for height in fam:  #height is an arbitrary name. any variable can be used
    print(height) 
```

## enumerate

```python
fam = [1.73, 1.68, 1.71, 1.89]
for index, height in enumerate(fam):
    print("index " + str(index) + ": " + str(height))
```

```python
for c in "family":
    print(c.capitalize())
```
